## 4. Missing Value Analysis

Check whether the dataset contains missing values and determine appropriate handling strategies.

```python
# Count missing values in each column to identify columns with missing values.
df.isnull().sum()
```

### Observation

The dataset contains a small number of missing values in the Postal Code column.

Since postal code is not critical for this analysis, records with missing postal codes will be removed.

```python
# Remove rows with missing postal codes
df = df.dropna(subset=["Postal Code"])
```

## 5. Duplicate Analysis

Identify and remove duplicate records to ensure data quality.

```python
df.isnull().sum()
```

```python
# Count duplicate rows to ensure data quality by removing duplicate records.
df.duplicated().sum()
```

## 6. Date Transformation

Convert date columns into datetime format to support time-based analysis.

```python
# Convert order date to datetime format
# Convert date columns into datetime format for time-series analysis.
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    dayfirst=True
)

# Convert ship date to datetime format
df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    dayfirst=True
)
```

## 7. Feature Engineering

Create additional time-related features for sales trend analysis.

```python
# Extract year from order date
df["Year"] = df["Order Date"].dt.year

# Extract month from order date
df["Month"] = df["Order Date"].dt.month

# Extract quarter from order date
df["Quarter"] = df["Order Date"].dt.quarter
```

## 8. Exploratory Data Analysis

Perform basic exploratory analysis to understand sales distribution.

```python
# Total Sales
df["Sales"].sum()
```

```python
# Sales by Region
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

```python
# Sales by Segment
df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
```

```python
# Top 10 states
df.groupby("State")["Sales"].sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
#Top 10 products
df.groupby("Product Name")["Sales"].sum()\
.sort_values(ascending=False)\
.head(10)
```
